```python
import pandas as pd
import numpy as np

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
```

```python
names = ['target', 'rbarba', 'gbarba', 'bbarba', 'rsapato',
         'gsapato', 'bsapato', 'rcamisa', 'gcamisa', 'bcamisa', 
        'rcalça', 'gcalça', 'bcalça','rcabelo', 
        'gcabelo', 'bcabelo']
```

```python
dataset = pd.read_csv('dados_treinamento.txt', header=None, names=names)
dataset
```

```python
dataset_test = pd.read_csv('dados_teste.txt', header=None, names=names)
dataset_test
```

**X contém todos os dados do dataset de treino, com exceção do campo target**

```python
X_train = np.array(dataset.drop('target', axis=1))
print(X_train)
```

**Y contém apenas o campo target do dataset de treino**

```python
y_train = np.array(dataset.target)
print(y_train)
```

**X contém todos os dados do dataset de teste, com excessão do campo target**

```python
X_test = np.array(dataset_test.drop('target', axis=1))
print(X_test)
```

**Y contém apenas o campo target do dataset de teste**

```python
y_test = np.array(dataset_test.target)
print(y_test)
```

**Aqui usa a arvore de decisão com critério de entropia (Feito o calculo = 0.859)**

```python
arvore = DecisionTreeClassifier(criterion='entropy')
arvore.fit(X_train, y_train)
```

```python
arvore.score(X_train, y_train)
```

**Aqui é feito o cálculo do ganho de informação. O maior ganho de informação é na posição 2 (gbarba)**

```python
arvore.feature_importances_
```

```python
target_names = np.array(['Marge', 'Homer'])
```

**Pega o atributo com maior ganho de informaçao e gera a árvore**

```python
from sklearn import tree
tree.plot_tree(arvore, feature_names=names, class_names=target_names, filled=True);
```

```python
y_pred = arvore.predict(X_test)

from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```

```python
y_pred
```

```python
print(classification_report(y_test, y_pred, target_names=target_names))
```
