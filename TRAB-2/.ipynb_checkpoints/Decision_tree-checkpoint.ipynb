```python
#Importando pandas
import pandas as pd
import numpy as np
from sklearn import tree
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import plot_confusion_matrix
```

```python
#Criando lista de names
names = ['target', 'rbarba', 'gbarba', 'bbarba', 'rsapato',
         'gsapato', 'bsapato', 'rcamisa', 'gcamisa', 'bcamisa', 
        'rcalça', 'gcalça', 'bcalça','rcabelo', 
        'gcabelo', 'bcabelo']
```

```python
#Dataset de treino
dataset_train = pd.read_csv('dados_treinamento.txt', header=None, names=names)
dataset_train.head(5)
```

```python
#Dataset de teste
dataset_test = pd.read_csv('dados_teste.txt', header=None, names=names)
dataset_test.head(5)
```

```python
#Valores do target
dataset_train.target.value_counts()

target_names = np.array(['Marge', 'Homer'])
```

```python
#Selecionando apenas 2 columnas para melhor visualização
dataset_train1 = dataset_train.loc[dataset_train.target.isin([0,1]),['rbarba','gbarba', 'target']]

dataset_test1 = dataset_test.loc[dataset_train.target.isin([0,1]),['rbarba','gbarba', 'target']]
```

```python
#Separando X e Y
X_train = dataset_train1.drop('target', axis=1)
y_train = dataset_train1.target

X_test = dataset_test1.drop('target', axis=1)
y_test = dataset_test1.target
```

```python
#Visualizando os dados de treino
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

ax.scatter(X_train['rbarba'], 
           X_train['gbarba'], 
           c=y_train)

plt.show()
```

```python
#Visualizando os dados de treino já separados
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

ax.scatter(X_train['rbarba'], 
           X_train['gbarba'], 
           c=y_train)

ax.plot([233.5,233.5], [0, 220], '--r')
ax.plot([260.0,235.5], [177, 177], '--r')

plt.show()
```

```python
# Criando o classificador
clf = tree.DecisionTreeClassifier()
```

```python
#Fazendo o fit com os dados de treino
clf = clf.fit(X_train, y_train)
```

```python
#Verificando o score
clf.score(X_train, y_train)
```

```python
#Visualizando essa árvore
fig, ax = plt.subplots(figsize=(6,6))

tree.plot_tree(clf, feature_names=names, class_names=target_names)

plt.show()
```

```python
#Fazendo a previsão e avaliando o erro
y_pred = clf.predict(X_test)

from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
#Visualizando esses dados de teste no gráfico

import matplotlib.pyplot as plt

fig, ax = plt.subplots()

ax.scatter(X_test['rbarba'], 
           X_test['gbarba'], 
           c=y_test)

#ax.plot([45.0,45.0], [0, 250], '--r')

plt.show()
```

```python
#Fazendo o fit com os dados de teste
clf = clf.fit(X_test, y_test)
```

```python
#Verificando o score
clf.score(X_test, y_test)
```

```python
#Visualizando essa árvore

fig, ax = plt.subplots(figsize=(12,12))

tree.plot_tree(clf, feature_names=names, class_names=target_names)

plt.show()
```

```python
#Treinando e testando para todo o database
```

```python
#Separando X e Y para todo o dataset
X_train = dataset_train.drop('target', axis=1)
y_train = dataset_train.target

X_test = dataset_test.drop('target', axis=1)
y_test = dataset_test.target
```

```python
#Criando o classificador e fazendo o fit
clf2 = tree.DecisionTreeClassifier().fit(X_train, y_train)
```

```python
#Verificando o score para base de treino
clf2.score(X_train, y_train)
```

```python
#Visualizando a árvore
fig, ax = plt.subplots(figsize=(6,6))

tree.plot_tree(clf2, feature_names=names, class_names=target_names)

plt.show()
```

```python
#Verificando o score para base de testes
clf2.score(X_test, y_test)
```

```python
#Fazendo a previsão
y_pred2 = clf2.predict(X_test)
```

```python
#Avaliando o modelo
confusion_matrix(y_test, y_pred2)
```

```python
#Matriz de confusão
matrix = plot_confusion_matrix(clf2, X_test, y_test)
matrix.ax_.set_title('Matriz de confusão', color='white')
plt.xlabel('Previsto', color='white')
plt.ylabel('Real', color='white')
plt.gcf().axes[0].tick_params(colors='white')
plt.gcf().axes[1].tick_params(colors='white')
plt.gcf().set_size_inches(10,6)
plt.show()
```

```python
#Porcentagens de acerto
print(classification_report(y_test, y_pred2, target_names=target_names))
```
